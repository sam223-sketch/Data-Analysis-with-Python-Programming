```python
'''seaborn Histogram'''

import numpy as np
import pandas as pd
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import Series

ds1 = randn(1000)
ds2 = randn(100)

plt.hist(ds1)
plt.savefig('image1.png')
plt.show()
```

```python
plt.hist(ds1, density = True, color='green', bins=15, alpha=0.5)
plt.hist(ds2, density = True, color='green', bins=15, alpha=0.5)
plt.savefig('image2.png')
plt.show()
```

```python
#sns
ds3 = randn(1000)
ds4 = randn(1000)
sns_plot = sns.jointplot(ds3, ds4)
sns_plot.savefig('image3.png')
```

```python
sns_plot2 = sns.jointplot(ds3, ds4, kind='hex')
sns_plot2.savefig('image4.png')
```

```python
#manually create KDE by summing the gaussian distribution

ds = randn(100)
sns.rugplot(ds)
plt.hist(ds, alpha=0.5)
plt.savefig('KDE1.png')

x_axes = np.linspace(ds.min()-1,ds.max()+1,50)
```

```python
# bandwidth creation 
bandwidth = ((4*ds.std()**5)/(3*len(ds)))**0.2
kernels = []
for point in ds:
    kernel = stats.norm(point, bandwidth).pdf(x_axes)
    kernels.append(kernel)
    
    kernel = kernel/kernel.max()
    kernel = kernel * 0.6
    plt.plot(x_axes, kernel, alpha = 0.5, color = "red")
    
plt.savefig('KDE2.png')

kde = np.sum(kernels, axis = 0)
kde_fig = plt.plot(x_axes, kde, color = 'green')
sns.rugplot(ds)
plt.suptitle('KDE Plot')
plt.savefig('KDE3.png')
```

```python
# using seaborn shortcut

kdefig = sns.kdeplot(ds)
fig = kdefig.get_figure()
fig.savefig('KDE4.png')
```

```python
mean = [0,0]
covarience = [[1,0], [0,100]]
ds = np.random.multivariate_normal(mean, covarience, 500)
dframe = pd.DataFrame(ds, columns = ['col1', 'col2'])

fig = sns.kdeplot(dframe).get_figure()
fig.savefig('kde4.png')

#shade 
fig2 = sns.kdeplot(dframe, shade=True).get_figure()
fig2.savefig('kde5.png')
```

```python
#shade 
fig2 = sns.kdeplot(dframe, shade=True).get_figure()
fig2.savefig('kde6.png')
```

```python
#bandwidth change
fig3 = sns.kdeplot(dframe, bw='silverman').get_figure()
fig3.savefig('kde7.png')
```

```python
fig4 = sns.jointplot('col1','col2', dframe, kind = 'kde')
fig4.savefig('kde7.png')
```
